### **2. Data Cleaning**

Untuk memastikan tidak ada trailing space lagi dalam nilai di tabel, maka akan dicari seluruh nilai unique dari semua kolom yang memiliki nominal "$".

```python
df['Units Sold'].unique()
```

```python
df['Units Sold'] = df['Units Sold'].str.strip()
```

Disini terlihat bahwa data 'Units Sold' dalam $ memiliki tipe data string, oleh karena itu akan kita ganti menjadi float.

```python
df['Units Sold'] = df['Units Sold'].str.replace('$','').str.replace(',','').astype('float')
```

Selanjutnya, akan dirubah tipe data pada kolom Manufacturing Price, Sale Price,	Gross Sales, Discounts, Sales, COGS, dan Profit menjadi float. Dengan langkah yang sama seperti sebelumnya terhadap kolom Units Sold


```python
df['Manufacturing Price'] = df['Manufacturing Price'].str.strip()
df['Manufacturing Price'] = df['Manufacturing Price'].str.replace('$','').str.replace(',','').astype('float')
```

```python
df['Sale Price'] = df['Sale Price'].str.strip()
df['Sale Price'] = df['Sale Price'].str.replace('$','').str.replace(',','').astype('float')
```

```python
df['Gross Sales'] = df['Gross Sales'].str.strip()
df['Gross Sales'] = df['Gross Sales'].str.replace('$','').str.replace(',','').astype('float')
```

Pada kolom Discounts terhadap value "-", mari kita lihat unique values dari kolom tersebut.

```python
df['Discounts'].unique()
```

Setelah melihat values dari kolom Discounts, value "-" dapat kita rubah menjadi 0.

```python
df['Discounts'] = df['Discounts'].str.strip()
```

```python
df['Discounts'] = df['Discounts'].str.replace('$','').str.replace(',','').str.replace('-',"0")
```

```python
df['Discounts'] = pd.to_numeric(df['Discounts']) # pd.to_numeric digunakan untuk mengatasi jika masih ada values yang bukan merupakan number
```

Mari kita lanjutkan merubah tipe data dari kolom Sales, COGS, dan Profit, kali ini menggunakan metode pd.to_numberic(df['column']), untuk mengatasi jika masih ada values dalam kolom tersebut yang bukan merupakan objek number.

```python
df['Sales'] = df['Sales'].str.strip()
df['Sales'] = df['Sales'].str.replace('$','').str.replace(',','').str.replace('-',"0")
df['Sales'] = pd.to_numeric(df['Sales'])
```

```python
df['COGS'] = df['COGS'].str.strip()
df['COGS'] = df['COGS'].str.replace('$','').str.replace(',','').str.replace('-',"0")
df['COGS'] = pd.to_numeric(df['COGS'])
```

```python
df['Profit'] = df['Profit'].str.strip()
df['Profit'] = df['Profit'].str.replace('$','').str.replace(',','').str.replace('-',"0")
```

Pada kolom Profit, terdapat value seperti "(4533.75)" yang berarti bahwa terjadi defisit, yang akan kita rubah menjadi nilai negatif menggunakan metode .apply()

```python
def numeric_negative(profit):
    if '(' in profit and ')' in profit:
        return -1 * pd.to_numeric(profit.replace('(', '').replace(')', ''))
    else:
        return pd.to_numeric(profit)

df['Profit'] = df['Profit'].apply(numeric_negative)
```

### **3. Data Normalization**

Setelah melihat data yang sudah di bersihkan, dapat dilakukan normaliasi dengan membagi df menjadi 5 tabel. Yaitu:

```python
countries = df[['Country']].copy() # .copy() digunakan agar tabel terpisah dapat di edit lagi jika perlu tanpa merubah value dalam df awal.

segment = df[['Segment']].copy()

discount = df[['Discount Band']].copy()

product_info = df[['Product','Manufacturing Price', 'Sale Price']].copy()

dates = df[['Month Name','Year','Month Number','Date']].copy()

sales_information = df[['Units Sold','Gross Sales','Discounts','Sales','COGS','Profit']].copy()
```

Selanjutnya, mari kita rubah semua tabel tersebut dalam bentuk csv.

```python
countries.to_csv('countries.csv', index=False)
segment.to_csv('segment.csv', index = False)
product_info.to_csv('product_info.csv', index=False)
sales_information.to_csv('sales_information.csv', index=False)
dates.to_csv('dates.csv', index=False)
discount.to_csv('discount.csv', index=False)
```
